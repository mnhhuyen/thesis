```python
#ICD codes for congestive heart failure patients
import pandas as pd
import pandasql as psql
```

```python
diag_item = pd.read_csv('/media/data3/biodataset/MIMIC_IV/MIMIC-IV-v3.0/physionet.org/files/mimiciv/3.0/hosp/d_icd_diagnoses.csv')
diag_item.head()
```

```python
cond_icd9 = (
    (diag_item['icd_code'].str[:3] == '428') |
    (diag_item['icd_code'].str[:5].isin(['39891', '40201', '40211', '40291', '40401', '40403', '40411', '40413', '40491', '40493'])) |
    (diag_item['icd_code'].str[:4].between('4254', '4259'))
)

cond_icd10 = (
    (diag_item['icd_code'].str[:3].isin(['I43', 'I50'])) |
    (diag_item['icd_code'].str[:4].isin(['I099', 'I110', 'I130', 'I132', 'I255', 'I420', 'I425', 'I426', 'I427', 'I428', 'I429', 'P290']))
)

filtered_diag_item_all = diag_item[cond_icd9 | cond_icd10]
filtered_diag_item_all
```

```python
filtered_diag_item_all.to_csv('list_chf_code.csv')
```

```python
filtered_diag_item_code = filtered_diag_item_all['icd_code'].tolist()
filtered_diag_item_code
```

```python
diagnoses = pd.read_csv('/media/data3/biodataset/MIMIC_IV/MIMIC-IV-v3.0/physionet.org/files/mimiciv/3.0/hosp/diagnoses_icd.csv')
diagnoses.head()
```

```python
chf_diag = diagnoses[diagnoses['icd_code'].isin(filtered_diag_item_code)]
chf_diag
```

```python
filtered_chf_diag = chf_diag.drop_duplicates(subset='hadm_id')
filtered_chf_diag
```

```python
filtered_chf_diag.to_csv('patients_with_CHF.csv', index=False)
```
